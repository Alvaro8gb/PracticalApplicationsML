```python
import sys
sys.path.append("./../my-staff")
from database import load_ds
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

```

```python
X, y, features_names, class_names = load_ds("all")
```

```python
pca = PCA(n_components=2)

reduced_features = pca.fit_transform(X)
```

```python
# Calculate the percentage of variance explained by each of the selected components
variance_explained = pca.explained_variance_ratio_

# Create a scree plot
plt.figure(figsize=(8, 5))
plt.bar(range(1, len(variance_explained) + 1), variance_explained, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1, len(variance_explained) + 1), variance_explained.cumsum(), where='mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# Extract the eigenvalues
eigenvalues = pca.explained_variance_

# Create the scree plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Component Number')
plt.ylabel('Eigenvalue')


plt.show()
```
