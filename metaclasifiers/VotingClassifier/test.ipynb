```python
import matplotlib.pyplot as plt
from sklearn.ensemble import VotingClassifier
import sys
sys.path.append("../../my-staff")
from utils import load_ds
```

# Cargar dataset

```python
X, y, features_names, class_names = load_ds("all")
X.shape
```

```python
X_train, y_train, _, _ = load_ds("train")
X_train.shape
```

```python
X_test, y_test, _, _ = load_ds("test")
X_test.shape
```

```python
from utils import grid_search_no_fss, SEED
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
def create_meta():
    log_clf = LogisticRegression(random_state=SEED)
    svm_clf = SVC(random_state=SEED)
    rf_clf = RandomForestClassifier(random_state=SEED)

    return VotingClassifier(
        estimators=[('lr', log_clf), ('svc', svm_clf), ('rf', rf_clf)],
        voting='soft'
    )
```

```python
model = create_meta()
```

```python

param_grid = {
    'lr__C': [0.1, 1, 10],  # Parámetros para LogisticRegression
    'svc__C': [0.1, 1, 10],  # Parámetros para SVC
    'rf__n_estimators': [10, 50, 100],  # Parámetros para RandomForestClassifier
    'voting': ['hard', 'soft']  # Modo de votación
}
```

```python
result = grid_search_no_fss(X_train, y_train, model, param_grid)
```

```python
result
```

```python
best_model = result.best_model
```

```python
hipers = best_model.get_params()
hipers
```

```python
features = result.selected_feature_names
features
```

# Training the model

```python
model = create_meta()
model.set_params(**hipers)
```

```python
model.fit(X_train[features], y_train)
```

# Evaluation

```python
from utils import eval_model
```

```python
print("Training score: ", model.score(X_train[features], y_train))
print("Test score: ", model.score(X_test[features], y_test))
```

```python
y_prob = model.predict_proba(X_test[features])[:,0]
y_pred = model.predict(X_test[features])
eval_model(y_true= y_test, y_pred=y_pred, y_prob=y_prob)
```
